```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
nadia=cv2.imread('D:/Courses/#Computer Vision/Computer-Vision-with-Python/DATA/Nadia_Murad.jpg',0)
denis=cv2.imread('D:/Courses/#Computer Vision/Computer-Vision-with-Python/DATA/Denis_Mukwege.jpg',0)
solvay=cv2.imread('D:/Courses/#Computer Vision/Computer-Vision-with-Python/DATA/solvay_conference.jpg',0)
```

```python
plt.imshow(denis,cmap='gray')
```

```python
plt.imshow(solvay,cmap='gray')
```

```python
face_cascade=cv2.CascadeClassifier('/media/abrar/DATA/Courses/#Computer Vision/Computer-Vision-with-Python/DATA/haarcascades/haarcascade_frontalface_default.xml')
```

```python
def detect_face(img):
    face_img=img.copy()
    face_rect=face_cascade.detectMultiScale(face_img)
    
    for (x,y,w,h) in face_rect:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255),10)
    return face_img
```

```python
result=detect_face(solvay)
```

```python
plt.imshow(result,cmap='gray')
```

```python
def adjusted_detect_face(img):
    face_img=img.copy()
    face_rects=face_cascade.detectMultiScale(face_img,scaleFactor=1.15,minNeighbors=7)
    
    for (x,y,w,h) in face_rects:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255),10)
    return face_img
```

```python
result=adjusted_detect_face(solvay)
```

```python
plt.imshow(result,cmap='gray')
```

```python
eye_cascade=cv2.CascadeClassifier('/media/abrar/DATA/Courses/#Computer Vision/Computer-Vision-with-Python/DATA/haarcascades/haarcascade_eye.xml')
```

```python
def detect_eyes(img):
    face_img=img.copy()
    eyes_rect=eye_cascade.detectMultiScale(face_img,scaleFactor=1.15,minNeighbors=7)
    
    for (x,y,w,h) in eyes_rect:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255),10)
    return face_img
```

```python
result=detect_eyes(nadia)
plt.imshow(result,cmap='gray')
```

```python
result=detect_eyes(denis)
plt.imshow(result,cmap='gray')
```

```python
cap=cv2.VideoCapture(0)

while True:
    
    ret,frame=cap.read(0)
    frame=detect_face(frame)
    
    cv2.imshow('face detection',frame)
    
    k=cv2.waitKey(1)
    if k==27:
        break
cap.release()
cv2.destroyAllWindows()
```

```python
cap=cv2.VideoCapture(0)

while True:
    
    ret,frame=cap.read(0)
    frame=detect_eyes(frame)
    
    cv2.imshow('eyes detection',frame)
    
    k=cv2.waitKey(1)
    if k==27:
        break
cap.release()
cv2.destroyAllWindows()
```
